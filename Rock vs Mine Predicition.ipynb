```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv(r'\Users\Sagar singh\jupter notebook\excell sheet\Copy of sonar data.csv',header = None)
```

```python
X = sonar_data.drop(columns = 60,axis =1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

training  and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape,Y_test.shape,Y_train.shape)
```

# model training --> logistic regression  

```python
model = LogisticRegression()
```

```python
# traiinng the logictic regression model with trainig data
model.fit(X_train,Y_train)
```

```python
# model evaluation 
```

```python
# accuracy on train data
X_train_prediction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on traning data: " ,traning_data_accuracy)
```

```python
# accuray on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data: " ,test_data_accuracy)
```
